```python
import pandas as pd
import numpy as np 
from matplotlib import pyplot as plt
import seaborn as sns

plt.style.use ('ggplot')
```

```python
df=pd.read_excel("/kaggle/input/datasets/ahmedosman220/candy-data/candyhierarchy2017.xlsx")
df.head(10)
```

```python
df.drop(columns='Unnamed: 113',inplace=True)
df.head()
```

```python
df.set_index('Internal ID',inplace =True)
df.head()
```

```python
df['Q1: GOING OUT?']=df['Q1: GOING OUT?'].fillna("I'd rather not say")
df['Q1: GOING OUT?'].unique()
```

```python
df['Q2: GENDER']=df['Q2: GENDER'].fillna("I'd rather not say")

df['Q2: GENDER'].unique()
```

```python
df['Q3: AGE']=pd.to_numeric(df['Q3: AGE'],errors='coerce')
df['Q3: AGE']=df['Q3: AGE'].where(df['Q3: AGE'] .between(1,100))
b=round(df['Q3: AGE'].mean(),1)
df['Q3: AGE']=df['Q3: AGE'].fillna(b)

df['Q3: AGE'].value_counts()
```

```python
df['Q4: COUNTRY']=df["Q4: COUNTRY"].fillna("unknowen").str.lower().str.replace (r'[^a-zA-Z]','',regex=True).str.strip()

dic1={'unitedstates':'USA','usa':'USA','us':'USA','unitedstatesofamerica':'USA','':'Unknowen','unitedkingdom':'UK','unitedstaes':'USA','subscribetodmuzonyoutube':'Unknowen','idontknowanymore':'Unknowen','fearandloathing':'Unknowen','ussa':'USA'}
dic2={'usausausa':'USA','america':'USA','unhingedstates':'USA','usofa':'USA','unitesstates':'USA','theunitedstates':'USA','uniedstates':'USA','earth':'Unknowen','europe':'Unknowen','theunitedstatesofamerica':'USA','unitestates':'USA','insanitylately':'Unknowen'}
dic3={'usahardtotellanymore':'USA','merica':'USA','usas':'USA','unitedsate':'USA','a':'Unknowen','can':'Canada','canae':'Canada','unitedsates':'USA','southafrica':'South Africa','usaa':'USA','Sovieatcanuckistan':'Sovieat canuckistan','namerica':'USA'}
dic4={'ipretendtobefromcanadabutiamreallyfromtheunitedstates':'USA','unitedstate':'USA','ahemamerca':'USA','ud':'UK','unitedststes':'USA','thenetherlands':'Netherlands','unitedstatss':'USA','endland':'UK','trumpistan':'USA','unitedstatea':'USA'}

df['Q4: COUNTRY']=df['Q4: COUNTRY'].replace(dic1).replace(dic2).replace(dic3).replace(dic4).str.title()

df['Q4: COUNTRY'].unique()
```

```python
df['Q5: STATE, PROVINCE, COUNTY, ETC']=df['Q5: STATE, PROVINCE, COUNTY, ETC'].fillna('Unknowen').str.lower().str.replace (r'[^a-zA-Z]',' ',regex=True).str.strip().str.title()
dict1={'Ca':'California','The Democratic People S Republic Of North Carolina':'North Carolina','Psychotic':'Unknowen','Londom':'london','Pennsylvania  Northampton Couny  Easton':'Pennsylvania  Northampton County  Easton','Quenec':'Quebec','Ur Mom':'Unknowen','A':'Unknowen','Usa':'USA','Cascadia':'Unkowen','Nope':'Unknowen','1':'Unknowen','48':'Unknowen',}



df['Q5: STATE, PROVINCE, COUNTY, ETC'] =df['Q5: STATE, PROVINCE, COUNTY, ETC'].astype(str).str.strip().str.title().replace(dict1)
df['Q5: STATE, PROVINCE, COUNTY, ETC'].unique()

```

```python
df['Q11: DAY']=df['Q11: DAY'].fillna('Unknowen')
df['Q11: DAY'].value_counts()
```

```python
df['Q10: DRESS']=df['Q10: DRESS'].fillna("I'd rather not to say")
df['Q10: DRESS'].value_counts()
```

```python
df['Q9: OTHER COMMENTS']=df['Q9: OTHER COMMENTS'].fillna("I don't have comments")
df['Q9: OTHER COMMENTS'].unique()
```

```python
df['Q7: JOY OTHER']=df['Q7: JOY OTHER'].fillna("I'd rether not to say")
df['Q7: JOY OTHER'].unique()
```

```python

x=df['Q2: GENDER'].value_counts()
val=x.values
num=x.index
color= sns.color_palette("muted6")
plt.pie(val,labels=num,colors=color,autopct='%1.1f%%',shadow=True)
plt.title('Who is eating all this candy ?')
plt.show()
```

```python
y=df['Q4: COUNTRY'].value_counts().head(5)
n=y.values
m=y.index
sns.barplot(x=m,y=n,palette="magma",edgecolor='white')
plt.xlabel('Name of Countries')
plt.ylabel('Number of Countries')
plt.title('Distribution of Country')
plt.show()
```

```python
q= [col for col in df.columns if col.startswith ('Q6 |')]
results={}
for col in q:
    count = (df[col]=='JOY').sum()
    results[col]=count 

result=pd.Series(results)
e=result.sort_values(ascending=False).head(15)
r=e.values
t=e.index

sns.barplot(x=r,y=t,palette="magma")
plt.title("The Most Loved Kind of Chocolate ")
plt.xlabel("KINDS OF CHOCOLATES")
plt.ylabel("NUMBER OF LOVERS")

plt.show()
```

```python
p=df['Q3: AGE']

sns.histplot(p,bins=20,kde=True,color="purple",edgecolor="white")
plt.title("Age Distribution")
plt.xlabel("Ages")
plt.ylabel("Number of People")

plt.show()
```

```python
k=df.loc[df['Q1: GOING OUT?'].str.contains('Yes'),'Q2: GENDER'].value_counts()
label=k.index
size=k.values
color=sns.color_palette("muted")
plt.pie(size,labels=label,colors=color,autopct='%1.1f%%',shadow=True,wedgeprops={'width': 0.6})
plt.title("The relationship between gender and going out to celebrate")
plt.show()
```
